### 1.2 Feature Engineering Steps

These approaches were found to perform more poorly during validation and were ultimately determined to be less suitable for capturing subtle emotional cues from text.

From this process, TF-IDF was found to be the most effective method, given it helped provide varying weights to common vs unique words and enabled the identification of bigrams that may otherwise be difficult for the final model to interpret such as "not good" or "super bad".

At this stage, a lemmatizer function was defined using NLTK's WordNetLemmatizer library to enhance the model's ability to generalise words by consolidating those words that were semantically similar. 

Through a series of trial and eror, the hyperparameters for the TF-IDF vectorizer such as n-gram range, vocab size, and min/max document freeuqncy were determined by observing how these affected the resulting validation performanc metrics. Once this was finalsised, the TF-IDF vectoriser was fit on the training data and used to transform both validation and testing sets.

## 1. Preprocessing Steps

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

os.chdir(r"C:\Projects\DM2025-Lab2-Exercise\dm-lab-2-private-competition")

#Loading initial datasets 
posts = pd.read_json("final_posts.json")
labels = pd.read_csv("emotion.csv")
ids = pd.read_csv("data_identification.csv")

#Handling json nested structure, given we want to deal with post_id directly
posts_expand = posts['root'].apply(lambda x: x['_source']['post'])
posts_df = pd.json_normalize(posts_expand)

#Checking data structure and column labels
#print(posts_df.columns)
#print(ids.columns)
#print(labels.columns)

#Merging posts with labels, then with data identification (ids)
df = posts_df.merge(labels, left_on="post_id", right_on="id", how="left").drop(columns=['id'])
df = df.merge(ids, left_on="post_id", right_on="id", how="left").drop(columns=['id'])

#Splitting into training and testing sets, dropping redundant columns
train_df = df[df['split'] == 'train'].reset_index(drop=True)
test_df = df[df['split'] == 'test'].reset_index(drop=True)

#Shuffling training data for later model training and ensuring stability
train_df = train_df.sample(frac=1, random_state=8).reset_index(drop=True)
```

```python
#Defining a function for word lemmatization to improve word recognition based on their contextual meaning
#to be implemented in the subsequent feature engineering stage
import re
import nltk
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
#nltk.download('punkt')
#nltk.download('wordnet')

def lemmatize_tokenizer(text):
    return [lemmatizer.lemmatize(tok) for tok in nltk.word_tokenize(text.lower())]
```
